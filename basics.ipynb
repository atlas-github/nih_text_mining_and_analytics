```python
# Define the list
my_list = [1, 2, 3, 4, 5]

# Access element by index
first_element = my_list[0]   # Accessing the first element
third_element = my_list[2]   # Accessing the third element

# Extract a subset of elements (slicing)
subset = my_list[1:4]        # Extracting elements from index 1 to 3 (2, 3, 4)

# Adding elements to the list
my_list.append(6)            # Adding an element to the end
my_list.extend([7, 8])       # Adding multiple elements to the end

# Inserting elements
my_list.insert(2, 2.5)       # Inserting an element at index 2

# Deleting elements
del my_list[3]               # Deleting the element at index 3
removed_element = my_list.pop(4)  # Removing and returning the element at index 4
my_list.remove(2.5)          # Removing the first occurrence of the value 2.5

# Length of list
list_length = len(my_list)   # Getting the length of the list

# Membership testing
is_three_in_list = 3 in my_list    # Checking if 3 is in the list
is_ten_in_list = 10 in my_list     # Checking if 10 is in the list
```

```python
# Define the set
my_set = {1, 2, 3, 4, 5}

# Adding elements to the set
my_set.add(6)                 # Adding a single element to the set
my_set.update([7, 8])         # Adding multiple elements to the set

# Removing elements from the set
my_set.remove(3)              # Removing an element from the set (raises KeyError if not found)
my_set.discard(9)             # Discarding an element from the set (no error if not found)
removed_element = my_set.pop() # Removing and returning an arbitrary element from the set

# Union of sets
another_set = {4, 5, 6, 7, 8, 9, 10}
union_set = my_set.union(another_set)  # Union of my_set and another_set

# Intersection of sets
intersection_set = my_set.intersection(another_set)  # Intersection of my_set and another_set

# Difference of sets
difference_set = my_set.difference(another_set)      # Difference of my_set and another_set

# Length of set
set_length = len(my_set)       # Getting the number of elements in the set

# Membership testing in set
is_two_in_set = 2 in my_set    # Checking if 2 is in the set
is_three_in_set = 3 in my_set  # Checking if 3 is in the set

# Printing the results
print("Set after additions:", my_set)
print("Removed element:", removed_element)
print("Union of sets:", union_set)
print("Intersection of sets:", intersection_set)
print("Difference of sets:", difference_set)
print("Length of the set:", set_length)
print("Is 2 in the set?", is_two_in_set)
print("Is 3 in the set?", is_three_in_set)
```

```python
# Define the dictionary
my_dict = {'name': 'John', 'age': 30}

# Access element by key
name = my_dict['name']       # Accessing the value associated with the key 'name'
age = my_dict['age']         # Accessing the value associated with the key 'age'

# Adding and updating key-value pairs
my_dict['city'] = 'New York'         # Adding a new key-value pair
my_dict['age'] = 31                  # Updating the value associated with the key 'age'

# Length of dictionary
dict_length = len(my_dict)           # Getting the number of key-value pairs

# Membership testing
is_name_in_dict = 'name' in my_dict  # Checking if 'name' is a key in the dictionary
is_country_in_dict = 'country' in my_dict  # Checking if 'country' is a key in the dictionary

# Printing the results
print("Name:", name)
print("Age:", age)
print("Updated dictionary:", my_dict)
print("Length of the dictionary:", dict_length)
print("Is 'name' a key in the dictionary?", is_name_in_dict)
print("Is 'country' a key in the dictionary?", is_country_in_dict)
```

```python
# Define the string
my_string = "Hello, World!"

# Accessing characters by index
first_char = my_string[0]       # Accessing the first character
sixth_char = my_string[5]       # Accessing the sixth character (index 5)

# Extract substrings (slicing)
substring = my_string[7:12]     # Extracting characters from index 7 to 11 ("World")

# Concatenation
new_string = my_string + " How are you?"  # Concatenating with another string

# Length
string_length = len(my_string)  # Getting the length of the string

# Membership testing
is_hello_in_string = "Hello" in my_string  # Checking if "Hello" is a substring
is_hey_in_string = "Hey" in my_string      # Checking if "Hey" is a substring

# Printing the results
print("First character:", first_char)
print("Sixth character:", sixth_char)
print("Substring (characters from index 7 to 11):", substring)
print("Concatenated string:", new_string)
print("Length of the string:", string_length)
print("Is 'Hello' in the string?", is_hello_in_string)
print("Is 'Hey' in the string?", is_hey_in_string)
```

```python
# Wine Quality Clustering: Cluster wines into distinct groups based on their chemical properties.

# Load the wine dataset using sklearn.datasets.load_wine.

import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import numpy as np

wine = load_wine()
X = wine.data
```

```python
# Apply K-means clustering using KMeans.
# Determine the optimal number of clusters using the elbow method or silhouette score.

inertia = []
silhouette_scores = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))
```

```python
# Elbow method plot
plt.figure(figsize=(10, 5))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Silhouette score plot
plt.figure(figsize=(10, 5))
plt.plot(K, silhouette_scores, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Optimal k')
plt.show()
```

```python
# Visualize the clusters and interpret the results.

optimal_k = 2
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title(f'Wine Clusters with k={optimal_k}')
plt.colorbar()
plt.show()
```

# 1200 pm: Parse and analyze data from Microsoft Excel



```python
import pandas as pd

#Load all 3 sheets as dataframes in Colab

df_orders = pd.read_excel("Sample - Superstore.xlsx", sheet_name = "Orders")
df_returns = pd.read_excel("Sample - Superstore.xlsx", sheet_name = "Returns")
df_people = pd.read_excel("Sample - Superstore.xlsx", sheet_name = "People")
```

```python
# What is the total sales by category and year?

import seaborn as sns
import matplotlib.pyplot as plt

# Extract year from Order Date
df_orders['Order Year'] = pd.DatetimeIndex(df_orders['Order Date']).year

# Group by Category and Year to calculate total sales and total profit
sales_by_category_year = df_orders.groupby(['Category', 'Order Year'])['Sales'].sum().reset_index()
profit_by_category_year = df_orders.groupby(['Category', 'Order Year'])['Profit'].sum().reset_index()

# Group by City to find the total sales per city
sales_by_city = df_orders.groupby('City')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False).head(10)

# Plot total sales by category and year
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_category_year, x='Order Year', y='Sales', hue='Category')
plt.title('Total Sales by Category and Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Category')
plt.show()
```

```python
# What is the total profit by category and year?

plt.figure(figsize=(12, 6))
sns.barplot(data=profit_by_category_year, x='Order Year', y='Profit', hue='Category')
plt.title('Total Profit by Category and Year')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.legend(title='Category')
plt.show()
```

```python
# Which cities have generated the highest sales?

plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_city, x='Sales', y='City', palette='viridis')
plt.title('Top 10 Cities by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('City')
plt.show()
```

```python
# What are the margins by segment?

# Group by Segment and sum the sales and profit
segment_summary = df_orders.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Calculate the profit margin for each segment
# Profit margin is calculated as (Total Profit / Total Sales) * 100
segment_summary['Profit Margin'] = (segment_summary['Profit'] / segment_summary['Sales']) * 100

# Plot the profit margins by segment
plt.figure(figsize=(10, 6))
sns.barplot(data=segment_summary, x='Segment', y='Profit Margin', palette='viridis')
plt.title('Profit Margin by Segment')
plt.xlabel('Segment')
plt.ylabel('Profit Margin (%)')
plt.show()
```

```python
# Which city has the highest returns?

# Perform a left join operation between orders and returns on 'Order ID'
merged_data = pd.merge(df_orders, df_returns, on='Order ID', how='left', indicator=True)

# Filter out the returned orders
returned_orders = merged_data[merged_data['_merge'] == 'both']

# Calculate the number of returns per city
returns_by_city = returned_orders['City'].value_counts().reset_index()
returns_by_city.columns = ['City', 'Returns']

# Identify the city with the highest number of returns
top_return_city = returns_by_city.iloc[0]

# Plot the top 10 cities with the highest returns
plt.figure(figsize=(12, 6))
sns.barplot(data=returns_by_city.head(10), x='Returns', y='City', palette='viridis')
plt.title('Top 10 Cities by Number of Returns')
plt.xlabel('Number of Returns')
plt.ylabel('City')
plt.show()

# Display the city with the highest number of returns
top_return_city
```

```python
# What is the name of the Customer with the Highest Sales in New York City?

# Filter the data for orders in New York City
nyc_orders = df_orders[df_orders['City'] == 'New York City']

# Group by Customer Name and sum the Sales
sales_by_customer_nyc = nyc_orders.groupby('Customer Name')['Sales'].sum().reset_index()

# Identify the customer with the highest total sales in New York City
top_customer_nyc = sales_by_customer_nyc.loc[sales_by_customer_nyc['Sales'].idxmax()]

# Display the name of the customer with the highest sales in New York City
top_customer_nyc
```

```python
# What Sub-Category did this Customer buy, and how much Sales and Profit did this Customer generate for the company?

# Filter the data to include only orders from this customer
top_customer_orders = nyc_orders[nyc_orders['Customer Name'] == top_customer_nyc['Customer Name']]

# Group by Sub-Category to see what the customer bought
sub_category_summary = top_customer_orders.groupby('Sub-Category').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Calculate the total sales and profit generated by this customer
total_sales = top_customer_orders['Sales'].sum()
total_profit = top_customer_orders['Profit'].sum()

# Display the sub-category summary and total sales and profit
sub_category_summary, total_sales, total_profit
```

# 09.15 am: Revise data structures and list methods

```python
# Function to find the sum of all numbers in a list
def find_sum(numbers):
    return sum(numbers)
```

```python
find_sum([1, 2, 3, 4, 5])
```

```python
# Function to remove duplicates from a list
def remove_duplicates(lst):
    return list(set(lst))
```

```python
remove_duplicates([1, 2, 2, 3, 4, 4, 5])
```

```python
# Function to sort a list of strings in alphabetical order
def sort_strings(strings):
    return sorted(strings)
```

```python
sort_strings(['banana', 'apple', 'orange', 'grape'])
```

```python
# Function to count the frequency of each word in a list and store in a dictionary
def count_word_frequency(words):
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count
```

```python
count_word_frequency(['apple', 'banana', 'apple', 'orange', 'banana', 'banana'])
```

```python
# Function to merge two dictionaries
def merge_dictionaries(dict1, dict2):
    return {**dict1, **dict2}
```

```python
merge_dictionaries({'a': 1, 'b': 2}, {'c': 3, 'd': 4})
```

```python
# Function to check if a key exists in a dictionary
def check_key_existence(dictionary, key):
    return key in dictionary
```

```python
check_key_existence({'a': 1, 'b': 2, 'c': 3}, 'b')
```

```python
# Function to find common elements between two sets
def find_common_elements(set1, set2):
    return set1.intersection(set2)
```

```python
find_common_elements({1, 2, 3, 4}, {3, 4, 5, 6})
```

```python
# Function to find unique elements that appear in both lists
def find_unique_elements(list1, list2):
    return set(list1).intersection(list2)
```

```python
find_unique_elements([1, 2, 3, 4], [3, 4, 5, 6])
```

```python
# Function to find the difference between two sets
def find_set_difference(set1, set2):
    return set1.difference(set2)
```

```python
find_set_difference({1, 2, 3, 4}, {3, 4, 5, 6})
```

```python
# Function to reverse a string
def reverse_string(input_str):
    return input_str[::-1]
```

```python
reverse_string('hello')
```

```python
# Function to count occurrences of each character in a string and store in a dictionary
def count_character_occurrences(input_str):
    char_count = {}
    for char in input_str:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count
```

```python
count_character_occurrences('hello')
```

```python
# Function to capitalize the first letter of each word in a string
def capitalize_first_letter(input_str):
    return input_str.title()
```

```python
capitalize_first_letter('hello world')
```

```python
# a for loop that generates a list of squares of numbers from 1 to 5:

squares = []

for x in range(1, 6):
    squares.append(x ** 2)

print(squares)
```

```python
#list comprehension
squares = [x ** 2 for x in range(1, 6)]
```

```python
import timeit

# Define the code for the for loop
for_loop_code = """
squares = []

for x in range(1, 6000):
    squares.append(x ** 2)
"""

# Define the code for the list comprehension
list_comp_code = """
squares = [x ** 2 for x in range(1, 6000)]
"""

# Measure the time taken for the for loop
for_loop_time = timeit.timeit(for_loop_code, number=1000)

# Measure the time taken for the list comprehension
list_comp_time = timeit.timeit(list_comp_code, number=1000)

print("Time taken for the for loop:", for_loop_time)
print("Time taken for the list comprehension:", list_comp_time)
```

```python
# Multiples of Three and Five: Write a list comprehension that generates a list of numbers from 1 to 100 that are multiples of three or five.

multiples_of_three_or_five = []

for num in range(1, 101):
    if num % 3 == 0 or num % 5 == 0:
        multiples_of_three_or_five.append(num)

print(multiples_of_three_or_five)
```

```python
#list comprehension
[n for n in range(1, 101) if (n % 3 == 0) or (n % 5 == 0)]
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

#search terms here
search_terms = ["Avastin", "bevacizumab", "아바스틴", "베바시주맙"]

headers = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Language': 'en-US,en;q=0.9',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive',
    'Referer': 'https://kmbase.medric.or.kr/master.php',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-User': '?1',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'sec-ch-ua': '"Not_A Brand";v="8", "Chromium";v="120", "Google Chrome";v="120"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
}

for k in range(0, len(search_terms)):

  i = 0

  while i >= 0:

    params = {
        'query': str(search_terms[k]),
        'src': '150',
        'pg_idx': '100',
        'vidx': '',
        'list': 'Summary',
        'ln': '20',
        'old_pg_idx': '100',
        'pg': str(i),
        'page_num': '1',
        'frmtime': '1546272000',
    }

    response = requests.get('https://kmbase.medric.or.kr/master.php', params=params, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    test_list = list(soup.find_all("div", {"class": "list-table-cell t-a-l"}))

    for j in range(0, len(test_list)):
      article_id = test_list[j].find("input", {'name': "sltRow[]"})['value']
      title = test_list[j].find('p', class_='km_tit').text.strip()
      author_main = test_list[j].find('p', class_='km_auth_main').text.strip()
      author_sub = test_list[j].find('p', class_='km_auth_sub').text.strip()

      journal_info_all = test_list[j].find_all('p', class_='km_add')

      # Check if there is at least one element with class 'km_add'
      if len(journal_info_all) > 1:
          # Get the text of the second element
          journal_info = journal_info_all[1].text.strip()
      else:
          # Set journal_info to an empty string or handle the case accordingly
          journal_info = ""

      search_term_list.append(params["query"])
      article_id_list.append(article_id)
      title_list.append(title)
      author_main_list.append(author_main)
      author_sub_list.append(author_sub)
      journal_info_list.append(journal_info)

    if len(list(soup.find_all("div", {"class": "list-table-cell t-a-l"}))) == 0:
      break

    i += 1
```

```python
#turn lists into dataframes
df1 = pd.DataFrame(zip(*[search_term_list, article_id_list, title_list, author_main_list, author_sub_list, journal_info_list]), columns = ["Search_Term", "Article_ID", "Title", "Author_Main", "Author_Sub", "Journal_Info"])
df1
```

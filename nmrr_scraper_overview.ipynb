# Scrape overview seach page


```python
from bs4 import BeautifulSoup
from urllib3.exceptions import InsecureRequestWarning
import pandas as pd
import requests
import warnings

article_links_list = []
article_titles_list = []
article_identifiers_list = []
article_authors_list = []
```

```python
for i in range(1, 2046):

  # Suppress only the InsecureRequestWarning from urllib3
  warnings.filterwarnings("ignore", category=InsecureRequestWarning)

  response = requests.get("https://nmrr.gov.my/research-directory?page=" + str(i), , data=json_dataverify = False)
  soup = BeautifulSoup(response.text, "html.parser")

  #get article links
  soup_links = soup.find_all("a", href=lambda href: href and "https://nmrr.gov.my/research-directory/" in href)
  soup_links = list(soup_links)
  article_links = [a_tag['href'] for a_tag in soup_links]
  article_links_list.extend(article_links)

  #get article titles
  soup_title = []

  for title in soup_links:
    title = title.get_text()
    soup_title.append(title)

  article_titles = [title.strip() for title in soup_title]
  article_titles_list.extend(article_titles)

  #get article identifiers
  article_identifier = []

  soup_article_identifier = soup.find_all("div", class_ = {"text-sm font-medium text-gray-500"})
  soup_article_identifier = list(soup_article_identifier)

  for article_identifier_number in soup_article_identifier:
    article_identifier_number = article_identifier_number.get_text()
    article_identifier.append(article_identifier_number)

  article_identifiers = [identifiers.strip() for identifiers in article_identifier]
  article_identifiers_list.extend(article_identifiers)

  #get article authors
  article_authors = []

  soup_article_author = soup.find_all("div", class_ = {"mt-1 text-sm text-gray-500"})
  soup_article_author = list(soup_article_author)

  for article_author in soup_article_author:
    article_author = article_author.get_text()
    article_authors.append(article_author)

  article_authors = [author.strip() for author in article_authors]
  article_authors_list.extend(article_authors)
```

```python
df_nmrr_research_directory = pd.DataFrame(zip(*[article_titles_list, article_links_list, article_identifiers_list, article_authors_list]), columns = ["article_title", "article_link", "article_identifier", "article_author"])
df_nmrr_research_directory
```

```python
#get current time
from datetime import datetime
from zoneinfo import ZoneInfo

#Gets the current date and time, based on Singapore's timezone
singapore_dt = datetime.now(tz=ZoneInfo("Asia/Singapore"))
time = singapore_dt.strftime('%d-%b-%Y %I:%M:%S %p')
time
```

```python
df_nmrr_research_directory.to_csv("df_nmrr_research_directory_"+str(time)+".csv")

from google.colab import files
files.download("df_nmrr_research_directory_"+str(time)+".csv")
```

# Scrape detailed results

```python
import pandas as pd

df_nmrr = pd.read_csv("df_nmrr_research_directory_"+str(time)+".csv")
df_nmrr
```

```python
df_nmrr["page_title"] = ""
df_nmrr["page_sub_title"] = ""
df_nmrr["page_trial"] = ""
df_nmrr["page_research_type"] = ""
df_nmrr["page_description"] = ""
df_nmrr["page_research_keyword"] = ""
df_nmrr["page_inclusion_criteria"] = ""
df_nmrr["page_exclusion_criteria"] = ""
df_nmrr["page_date_study"] = ""
df_nmrr["page_date_actual_completed"] = ""
df_nmrr["study_duration"] = ""
```

```python
import requests

for i in range(0, len(df_nmrr)):
  response = requests.get(df_nmrr["article_link"][i], verify = False)
  soup = BeautifulSoup(response.text, "html.parser")

  #start with empty lists
  page_title_list = []
  page_sub_title_list = []
  page_trial_list = []
  page_research_type_list = []
  page_description_list = []
  page_research_keyword_list = []
  page_inclusion_criteria_list = []
  page_exclusion_criteria_list = []
  page_date_study_list = []
  page_date_actual_completed_list = []
  page_study_duration_list = []

  soup_page_title_list = soup.find("h1", class_ = {"text-2xl font-bold text-gray-900 sm:text-4xl py-1"}).get_text(strip=True)
  soup_page_sub_title_list = soup.find("h3", class_ = {"text-lg leading-6 font-medium text-gray-900"}).get_text(strip=True)
  soup_page_trial_list = soup.find("p", class_ = {"mt-1 max-w-2xl text-sm text-gray-500"}).get_text(strip=True)
  soup_page_research_type_list = soup.find("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"}).get_text(strip=True)
  soup_page_description_list = soup.find_all("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"})
  soup_page_research_keyword = soup.find_all("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"})
  soup_inclusion_criteria = soup.find_all("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"})
  soup_exclusion_criteria = soup.find_all("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"})
  soup_date_study = soup.find_all("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"})
  soup_date_actual_completed_study = soup.find_all("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"})
  soup_study_duration_study = soup.find_all("dd", class_ = {"mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2"})

  # Check if there is at least one element with class 'mt-1 text-sm text-gray-900 sm:mt-0 sm:col-span-2'
  if len(soup_page_description_list) > 1:
    #get the text of the second element:
    soup_page_description_list = soup_page_description_list[1].text.strip()
  else:
    soup_page_description_list = ""

  if len(soup_page_research_keyword) > 1:
    #get the text of the second element:
    soup_page_research_keyword = soup_page_research_keyword[4].text.strip()
  else:
    soup_page_research_keyword = ""

  if len(soup_inclusion_criteria) > 1:
    #get the text of the second element:
    soup_inclusion_criteria = soup_inclusion_criteria[5].text.strip()
  else:
    soup_inclusion_criteria = ""

  if len(soup_exclusion_criteria) > 1:
    #get the text of the second element:
    soup_exclusion_criteria = soup_exclusion_criteria[6].text.strip()
  else:
    soup_exclusion_criteria = ""

  if len(soup_date_study) > 1:
    #get the text of the second element:
    soup_date_study = soup_date_study[7].text.strip()
  else:
    soup_date_study = ""

  if len(soup_date_actual_completed_study) > 1:
    #get the text of the second element:
    soup_date_actual_completed_study = soup_date_actual_completed_study[8].text.strip()
  else:
    soup_date_actual_completed_study = ""

  if len(soup_study_duration_study) > 1:
    #get the text of the second element:
    soup_study_duration_study = soup_study_duration_study[9].text.strip()
  else:
    soup_study_duration_study = ""

  page_title_list.append(soup_page_title_list)
  page_sub_title_list.append(soup_page_sub_title_list)
  page_trial_list.append(soup_page_trial_list)
  page_research_type_list.append(soup_page_research_type_list)
  page_description_list.append(soup_page_description_list)
  page_research_keyword_list.append(soup_page_research_keyword)
  page_inclusion_criteria_list.append(soup_inclusion_criteria)
  page_exclusion_criteria_list.append(soup_exclusion_criteria)
  page_date_study_list.append(soup_date_study)
  page_date_actual_completed_list.append(soup_date_actual_completed_study)
  page_study_duration_list.append(soup_study_duration_study)

  df_nmrr.loc[i, "page_title"] = page_title_list
  df_nmrr.loc[i, "page_sub_title"] = page_sub_title_list
  df_nmrr.loc[i, "page_trial"] = page_trial_list
  df_nmrr.loc[i, "page_research_type"] = page_research_type_list
  df_nmrr.loc[i, "page_description"] = page_description_list
  df_nmrr.loc[i, "page_research_keyword"] = page_research_keyword_list
  df_nmrr.loc[i, "page_inclusion_criteria"] = page_inclusion_criteria_list
  df_nmrr.loc[i, "page_exclusion_criteria"] = page_exclusion_criteria_list
  df_nmrr.loc[i, "page_date_study"] = page_date_study_list
  df_nmrr.loc[i, "page_date_actual_completed"] = page_date_actual_completed_list
  df_nmrr.loc[i, "study_duration"] = page_study_duration_list
```
